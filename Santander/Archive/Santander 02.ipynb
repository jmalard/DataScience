```python
dados2 = pd.DataFrame(rescaled_reducedX,columns=['Col1','Col2','Col3','Col4'])
```

```python
sns.boxplot(data=dados2, orient='v')
```

```python
mean_Col2 = rescaled_reducedX[:,1].mean()
mean_Col3 = rescaled_reducedX[:,2].mean()
mean_Col4 = rescaled_reducedX[:,3].mean()
```

```python
############# criar função para substituir outliers pela média ################
```

```python
rescaled_reducedX[:,1] = list(map(lambda x : mean_Col2 if x > 0.4 else x,rescaled_reducedX[:,1]))
rescaled_reducedX[:,2] = list(map(lambda x : mean_Col3 if x > 0.5 else x,rescaled_reducedX[:,2]))
rescaled_reducedX[:,3] = list(map(lambda x : x if (x>0.28 and x<0.45) else mean_Col4,rescaled_reducedX[:,3]))
```

```python
#Machine Learning
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
# Definindo os valores para o número de folds
num_folds = 10
seed = 7

# Preparando a lista de modelos
modelos = []
modelos.append(('LR', LogisticRegression()))
modelos.append(('LDA', LinearDiscriminantAnalysis()))
modelos.append(('NB', GaussianNB()))
modelos.append(('KNN', KNeighborsClassifier()))
modelos.append(('CART', DecisionTreeClassifier()))
modelos.append(('SVC', SVC()))
modelos.append(('XGBoost', XGBClassifier()))

# Avaliando cada modelo em um loop
resultados = []
nomes = []

for nome, modelo in modelos:
    kfold = KFold(n_splits = num_folds, random_state = seed)
    cv_results = cross_val_score(modelo, rescaled_reducedX, y, cv = kfold, scoring = 'accuracy')
    resultados.append(cv_results)
    nomes.append(nome)
    msg = "%s: %f (%f)" % (nome, cv_results.mean(), cv_results.std())
    print(msg)
```
