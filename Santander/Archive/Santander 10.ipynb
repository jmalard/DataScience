Eliminando Duplicados

A funo abaixo utiliza o mtodo Unique do Pandas para identificar as variveis que possuem valores constantes. 
Estas variveis so colocadas em uma lista para posteriormente serem eliminadas.
Variveis com valores constantes podem comprometer o treinamento.

Documentao consultada:
https://machinelearningmastery.com/basic-data-cleaning-for-machine-learning/
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.unique.html
    

```python
def identify_constant_features(dataframe):
    count_uniques = dataframe.apply(lambda x: len(x.unique()))
    print(count_uniques)
    constants = count_uniques[count_uniques == 1].index.tolist()
    return constants
```

```python
constant_features_train = set(identify_constant_features(dados1))
```

```python
constant_features_train
```

```python
print('Existem {} variveis constantes.'.format(
        len(constant_features_train)))
```

```python
#Eliminando as colunas de valores constantes 
dados1.drop(constant_features_train, inplace=True, axis=1)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
#Separando o dataset em dois para separar o label e eliminando as colunas TARGET 
y_balanced_dados1 = balanced_dados1['TARGET'].values
x_balanced_dados1 = balanced_dados1.drop(['TARGET'], axis=1).values
#x_balanced_dados1 = balanced_dados1.drop(['ID'], axis=1).values
```

```python
#Eliminando as colunas de valores constantes 
dados2.drop(constant_features_train, inplace=True, axis=1)
```

```python
#Separando a coluna de IDs dos novos dados para executar o treinamento
ids_dados2 = dados2['ID']
#x_dados2 = dados2.drop(['ID'], axis=1).values
x_dados2 = dados2.values
```

```python
#Separando o conjunto de dados em dados de treino e teste ( 70% treino e 30% teste)
x_fit, x_eval, y_fit, y_eval= train_test_split(x_balanced_dados1, y_balanced_dados1, test_size=0.3)
```

```python
#Chacando os tamanhos
print([x_fit.shape,x_eval.shape,y_fit.shape,y_eval.shape]); 
```

```python
# Definindo os valores para o nmero de folds
num_folds = 10
seed = 7

# Preparando a lista de modelos
modelos = []
modelos.append(('LR', LogisticRegression()))
modelos.append(('LDA', LinearDiscriminantAnalysis()))
modelos.append(('NB', GaussianNB()))
modelos.append(('KNN', KNeighborsClassifier()))
modelos.append(('CART', DecisionTreeClassifier()))
modelos.append(('SVC', SVC()))
modelos.append(('XGBoost', XGBClassifier()))

# Avaliando cada modelo em um loop
resultados = []
nomes = []

# Avaliando modelos com Cross Validation e mostranto o melhor resultado
for nome, modelo in modelos:
    kfold = KFold(n_splits = num_folds, random_state = seed)
    cv_results = cross_val_score(modelo, x_fit, y_fit, cv = kfold, scoring = 'accuracy')
    resultados.append(cv_results)
    nomes.append(nome)
    msg = "%s: %f (%f)" % (nome, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
#Selecionado o modelo de melhor performance no teste de Cross Validation
modelo = XGBClassifier()
```
